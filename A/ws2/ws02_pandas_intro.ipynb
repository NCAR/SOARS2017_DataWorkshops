## Pandas in 2 minutes
[Pandas](http://pandas.pydata.org/) is a standard entry point into data manipulation in Python.  It provides a set of standard tools for loading, manipulating and geting basic statistical information about your data.  The documentation for Pandas is excellent and their [quick tutorial](http://pandas.pydata.org/pandas-docs/stable/10min.html#minutes-to-pandas) is highly recommended.

In essence Pandas provides you the ability to:

1. load data from various sources (CSV, databases, Web/HTTP, etc.),
2. manipulate, reshape and recompose the loaded data (e.g. remove bad data, filter data),
3. do basic statistical analysis on the data,
4. store the data back down into a datasource (CSV, database, etc.).

You will find Pandas a bundle of joy once you get the hang of it!

It might be best to think as Pandas in conjunction with and along side [Numpy](http://www.numpy.org) which provides a lot more functionality for manipulating large matrices of numeric data.  See Pandas as a complementary toolkit to Numpy -- and perhaps the library of choice when _beginning_ your data import and munging.

## Dataframes Primer

While Pandas has a lot to offer for your data munging and analysis needs, the core structure you're going to be dealing with is a [**Dataframe**](http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dataframe).  In essence, a Dataframe is not unlike a vector, or matrix ... but it provides a much more powerful interface to the _operations_ of either, so as such it is difficult to call it one or the other.  If you've ever used R or Matlab, Dataframes will not be a foreign concept.

Setting up Pandas in the NB environment just requires an import of Pandas, and while we're at it we're going to import [matplotlib](http://matplotlib.org) which will allow us to plot data and do other visualizations with our data.

## Inspecting a Dataframe
Now that we've assigned the Datafram `df` to the results of the CSV file -- you can take a look and see what is in it:

````python
df
````

### Reducing Dataframe output - some convenience functions `head()` and `tail()`

Notice that Jupyter reduces your output for you, showing some rows, `...` and then the last rows.  Say you did not want to see anything but the first few rows of the beginning or end.  Use `head()` and `tail()` respsectively.  If you don't supply an optional parameter, you get the first or last 5 lines, otherwise whatever number you place as a parameter will show you that number of lines.

````python
df.head()
````

````python
df.head(17) # the first 17 lines
````

````python
df.tail(9) # the last 9!
````

### Dataframe data access

Accessing a dataframe is not unlike accessing a list -- the basic form of slicing the dataframe __by rows__ looks like this:

    df[start_index:end_index:skip_size]

Let's take a look:

````python
df[741:742] # slicing by rows outputing only row index 741 (row number 742)
````

````python
df[1:20:2]  # slicing by rows skipping every 2nd item from 1 to 20
````

### More complex slicing

Now what gets returned when we use this form is another Dataframe. Accessing the rows and columns (also known as _indices_) might also be done a different way if you need a more complex slice.  Say you wanted the first and last 20 items in your data frame.  We can simply use [`pandas.concat(list_of_slices)`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.concat.html#pandas.concat)  to get the data.

````python
pd.concat([df[1:20], df[-20:]])  # could you do the same with head and tail? Try it!
````

### Slicing by filters
Say now you wanted to find all the data from a certain month -- let's pick on April.  We can a a little logic into our selection with a criteria -- namely `df['month']==2`.  Our result looks  as such:

````python
df[df['month']==4].head() # only April .. and just the first 5 lines with head() by default
````

What if you want more  than one filter, say you want just the first 10 days of April:

````python
df[(df['month']==4) & (df['day']<=10)]
````

### Shaping and filtering Dataframes further

````python
df[(df['month']>4) & (df['month']<10) & (df['raintot']>100)] # summer month days with more than 100 raintot
````

### DataFrame statistics with `describe()`

As we can see so far, DataFrames are awesome.  They simplify getting and accessing data that would likely end up in things like tables, but we get a lot of other nice things with DataFrames.  One thing we haven't seen yet is any _summary_ data for the file we've loaded.  

What if we want a summary of the data in our DataFrame?  This is made super-easy with [`DataFrame.describe()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.core.groupby.DataFrameGroupBy.describe.html#pandas.core.groupby.DataFrameGroupBy.describe).  This method has a lot to offer and you will be served well to read the documentation for it fully.  You will notice we get nice inter-quartile range data as well as minimum, maximum and standard deviation data.

````python
df.describe()
````

We can also restrict the data being summarized like in the example below, where we just want to summarize the min, max and mean of just the `lat` and `lon` columns.

````python
df.describe().loc[['min','mean','max'],['lat', 'lon']]
````

## Data Inquiries

There are many convenience functions in Pandas, only a few of which will be addressed here, but a let's explore a few questions that might drive an initial inquiry of data.

### Finding the $n$ largest and smallest values in a DataFrame

* If you want to find the top smallest or largest values in a DataFrame, use [`DataFrame.nsmallest()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.nsmallest.html#pandas.DataFrame.nsmallest) or [`DataFrame.nlargest()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.nlargest.html#pandas.DataFrame.nlargest).

````python
df.nsmallest(10,['raintot'])
````

````python
df.nlargest(5,['tmaxavg'])
````

### Sorting values 

Sorting data in your output DataFrame is a very useful thing to be able to do.  Often you'll want to just print some data in a sorted order or maybe even explore along some dimension and just take a peek at the data.  Let's take a look at the [`DataFrame.sort_values()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html#pandas.DataFrame.sort_values).

````python
df[df['month']==1].sort_values(['day','raintot'])
````

### Counting things

The [`DataFrame.count()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.count.html#pandas.DataFrame.count) is a great way to get off the ground with understanding the shape of your data.

````python
obs_per_month = df.loc[:,['month','day']].groupby(['month']).count()
print(obs_per_month)
````

````python
obs_per_month.mean()[0]
````

````python
print(obs_per_month[obs_per_month.day > obs_per_month.mean()[0]]) # this shouldn't be this way
# print obs_per_month[obs_per_month.day>obs_per_month.mean()]
````

## Basic Plotting

Plotting in Pandas is made simple with the [`DataFrame.plot()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html#pandas.DataFrame.plot) method.  While not giving you the full control that [matplotlib](http://matplotlib.org/) gives you, the plots for the basic case are outstanding and should be the first tool to consider when trying to visualize your data.

Use `plot()` to get a a basic view of your data like in the example below:

````python
df.plot()
````

Notice the `plot()` method will try to plot every column variable in your data, and **the default plot is a _line plot_**.

Now let's plot just one column (_raintot_) ... and also filter the data, since we apparently have a lot of "negative" rain totals (bad data).

### Line Plots

````python
df[df['raintot']>=0]['raintot'].plot()
````

There are a few other ways to do the same thing!

````python
df[df['raintot']>=0]['raintot'].plot(kind='line')
````

````python
df[(df['raintot']>=0)].sort_values(['month','day'],ascending=True)['raintot'].plot(kind='line')
````

### Scatterplots

Plotting with scatter plots is also very easy with the `kind` parameter ... just remember to make sure you're plotting _something_ (_x_) against _something_ (_y_).

````python
df[df['raintot']>=0].plot('raintot','tmaxavg',kind='scatter')
````

````python
df[df['raintot']>=0].plot('month','raintot',kind='scatter')
````

### Histograms

````python
df[df['raintot']>=0].plot('month','raintot',kind='hist')
````

### And many more in the cast ...

There are more plots that can be created with this method ... [see the documentation]() for more information to find if this very convenient method can make life easier for your basic needs.

### Exporting plots

There is a simple pattern to remember when exporting plots:

1. save the plot to a variable (e.g. `f = df.plot()`)
2. use `f.figure.savefig` ([see more on the documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.savefig))

You should consider exporting to 300dpi since this amount of resolution is sufficient for print media (e.g. journal articles), web, your own notebooks and even posters!

````python
f = df[df['raintot']>=0].plot('raintot','tmaxavg',kind='scatter')
f.figure.savefig('sample.tif',dpi=300)
````

### Plotting more complex data ranges

We'll combine some more complex queries of our data to get some more interesting plots.

````python
print(df[df.raintot>0] \
        .loc[:,['month','raintot']] \
        .groupby(['month']).sum())
        
df[df.raintot>0].loc[:,['month','raintot']].groupby(['month']).sum().plot()
````

````python
f = df[df.raintot>0].loc[:,['month','raintot']].groupby(['month']).sum().plot()
f.figure.savefig('rain_v_month_1981.png',dpi=300)
````
